```python
def ReadContacts(filename):
    "Reads the contatcs."
    allc=[]
    with open(filename, 'r') as f:
        for line in f:
            if 'contacts' in line:
                for line in f:
                    c = line.split()
                    if 'rep' in line:
                        break
                    else:
                        allc.append(c)
    return(allc)

def DFFromContacts(file):
    "Reads the contacts into a pandas dataframe."
    lines = ReadContacts(file)
    df = pd.DataFrame(lines)
    df.rename(columns = {1: "c1", 
                         2: "c2",
                         3:"dist",
                         4:"fc"}, 
                        inplace = True)
    
    count = len(df.index)
    print(count)
    
    df.dist = df.dist.astype(float)
    df.c1 = df.c1.astype(int)
    df.c2 = df.c2.astype(int)
    df.fc = df.fc.astype(float)

    return(df)

def MatrixFromContacts(df):
    "This reads the contatcs dataframe"
    matrix = df.pivot_table(index='c1',columns='c2',values='dist',fill_value=0)   
    
    indexres = np.arange(1,119)
    indexresD1 = np.arange(1,60)
    indexresD2 = np.arange(60,119)

    matrixc1_inter = matrix.reindex(indexresD1[::-1],fill_value=0.0, axis='index')
    matrixcc_inter = matrixc1_inter.reindex(indexresD2,fill_value=0.0, axis='columns')
    matrix_intra1 = matrixc1_inter.reindex(indexresD1,fill_value=0.0, axis='columns')
    matrix_intra = matrix.reindex(indexresD2[::-1],fill_value=0.0)
    matrix_intra2 = matrix_intra.reindex(indexresD2,fill_value=0.0, axis='columns')
    matrixc1 = matrix.reindex(indexres[::-1],fill_value=0.0)
    matrixcc = matrixc1.reindex(indexres,fill_value=0.0, axis='columns')
    
    matrixc1_inter = matrix.reindex(indexresD1[::-1],fill_value=0.0, axis='index')
    
    return(matrixcc)
```

```python
def Matrix(file,N,cutoff=145):
    """Get the matrix of contatcs. This returns the full NxN matric for N residues, the interface between 
    one monmer of the dimer (N/2 residues) and both intrachain matrices.
    """
    lines = ReadContacts(file)
    df = pd.DataFrame(lines)
    df.rename(columns = {1: "c1", 
                         2: "c2",
                         3:"dist",
                         4:"fc"}, 
                        inplace = True)
    
    count = len(df.index)
    print(count)
    
    df.dist = df.dist.astype(float)
    df.c1 = df.c1.astype(int)
    df.c2 = df.c2.astype(int)
    df.fc = df.fc.astype(float)
    df = df.loc[df.fc != 0.000]
    
    matrix = df.pivot_table(index='c1',columns='c2',values='dist',fill_value=0)
   # matrixoff = df.pivot_table(index='c1',columns='c2',values='fc',fill_value=0)
    
    #N is the number for one monomer
    indexres = np.arange(1,N+1)
    # if you want to zoom in to the binding domains and the core in contact with 
    #the binding domain, which you choose to be up to residue 145 in each monomer
    indexresD1 = np.arange(1,cutoff)
    indexresD2 = np.arange(360,360+(cutoff-1))
    
    #select the part of the contacts 
    matrixc1_inter = matrix.reindex(indexresD1[::-1],fill_value=0.0, axis='index')
    matrixcc_inter = matrixc1_inter.reindex(indexresD2,fill_value=0.0, axis='columns')
    matrix_intra1 = matrixc1_inter.reindex(indexresD1,fill_value=0.0, axis='columns') #monomer A
    matrix_intra = matrix.reindex(indexresD2[::-1],fill_value=0.0)
    matrix_intra2 = matrix_intra.reindex(indexresD2,fill_value=0.0, axis='columns') #monomer B
    matrixc1 = matrix.reindex(indexres[::-1],fill_value=0.0)
    matrixcc = matrixc1.reindex(indexres,fill_value=0.0, axis='columns') #matrix for all residues
    
    matrixc1_inter = matrix.reindex(indexresD1[::-1],fill_value=0.0, axis='index')
    matrixcc_inter = matrixc1_inter.reindex(indexresD2,fill_value=0.0, axis='columns') #just the interface between monomers
    
    
    return(matrixcc,matrixcc_inter,matrix_intra1,matrix_intra2)

def ReadContactTxtFiles(file,N):
    df = pd.read_csv( file,sep='\s+', header=None)
    df.rename(columns = {1: "c1", 
                         2: "c2",
                         3:"dist",
                         4:"fc"}, 
                        inplace = True)
    
    count = len(df.index)
    print(count)
    
    df.dist = df.dist.astype(float)
    df.c1 = df.c1.astype(int)
    df.c2 = df.c2.astype(int)
    df.fc = df.fc.astype(float)
    df = df.loc[df.fc != 0.000]
    
    matrix = df.pivot_table(index='c1',columns='c2',values='dist',fill_value=0)
   # matrixoff = df.pivot_table(index='c1',columns='c2',values='fc',fill_value=0)
    
    indexres = np.arange(1,N+1)
    indexresD1 = np.arange(1,145)
    indexresD2 = np.arange(360,144+360)

    matrixc1_inter = matrix.reindex(indexresD1[::-1],fill_value=0.0, axis='index')
    matrixcc_inter = matrixc1_inter.reindex(indexresD2,fill_value=0.0, axis='columns')
    matrix_intra1 = matrixc1_inter.reindex(indexresD1,fill_value=0.0, axis='columns')
    matrix_intra = matrix.reindex(indexresD2[::-1],fill_value=0.0)
    matrix_intra2 = matrix_intra.reindex(indexresD2,fill_value=0.0, axis='columns')
    matrixc1 = matrix.reindex(indexres[::-1],fill_value=0.0)
    matrixcc = matrixc1.reindex(indexres,fill_value=0.0, axis='columns')
    
    matrixc1_inter = matrix.reindex(indexresD1[::-1],fill_value=0.0, axis='index')
    matrixcc_inter = matrixc1_inter.reindex(indexresD2,fill_value=0.0, axis='columns')
    
    
    return(matrixcc_inter,matrix_intra1,matrix_intra2)

def MatrixRvsS(onlyS, onlyR,name):
    """Read the test files with only R and S contacts."""
    fig, ax = plt.subplots(figsize=(5,5))
    onlyS[onlyS > 0] = 10
    onlyR[onlyR > 0] = 10
    """Makes the actual seaborn heatmap"""
    sns.heatmap(onlyS,ax=ax,cmap=cmapblue,cbar=False,linewidths=.1, linecolor='grey') 
    sns.heatmap(onlyR,ax=ax,cmap=cmapred,cbar=False,linewidths=.1,linecolor='grey') 
    """Sets the tick parameters"""
    ax.tick_params(
        axis='both',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=True,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelbottom=True,
        right=False, left=True, labelleft=True,labelsize='small',
        width = 0.5,
        length = 3, color='k',direction ='in')

    for _, spine in ax.spines.items():
        """Sets the appearance of box lines around the plot"""
        spine.set_visible(True)
        spine.set(linewidth=1)
        """Sets the lines seperating the two monomers (specific for lacI!!!)"""
        #ax.axvline(x=59,color='k', linewidth=0.5)
        #ax.axhline(y=59,color='k', linewidth=0.5)

    
    """Sets axes equal"""
    ax.set_aspect("equal")

    plt.tight_layout()
    fig.savefig('/Users/mallu899/Desktop/'+'{}'.format(name)+'_cmap_all.png',dpi=1200,tight_layout=True)
```
